Eliminamos valores nulos

```python
data1 =data.fillna(method="bfill")
data1
```

```python
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
data1['MP'] =data1['MP'].fillna(method="ffill")
data1['FP'] =data1['FP'].fillna(method="ffill")
data1['Otros'] =data1['Otros'].fillna(method="ffill")
data1['Poliza'] =data1['Poliza'].fillna(method="ffill")
data1 = data1.drop(['Otros'], axis=1)
data1
```

```python
fig = plt.figure(figsize =(15, 8))
data1.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

##Obtenemos la desviación estándar para eliminar outliers en el df

```python
y=data1
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
data2= data1[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data2
```

```python
columnas_a_eliminar = ['Fecha', 'Folio', 'UUID', 'RFC', 'Proveedor', 'TIPO GASTO', 'Descripción', 'MP', 'Tipo','Status', 'Poliza']
data2 = data1.drop(columnas_a_eliminar, axis=1)
data2
```

```python
data2.to_csv("data2.csv")
```

```python
data_clean=data2.copy()
data_clean=data_clean.fillna(round(data2.mean(),1))
data_clean
```

```python
data_clean.to_csv("data_clean.csv")
```

```python
valores_nulos = data_clean.isnull().sum()
valores_nulos
```

```python
fig2 = plt.figure(figsize =(15, 8))
data_clean.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe DE")
plt.show() #dibujamos el diagrama
```

##Cuantiles para sustituir outliers en el df

```python
y=data_clean

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
data3_iqr= data_clean[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
data3_iqr.to_csv("data3_iqr.csv")
```

```python
folio = data['Folio']
Datos_limpios = pd.concat([folio, data3_iqr], axis=1)
Datos_limpios
```

```python
Datos_limpios.to_csv("Datos_limpios.csv")
```
