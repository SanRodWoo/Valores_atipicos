```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data= pd.read_excel('/content/Gastos y costos 20-23.xlsx',sheet_name = '2020',header=5)
```

```python
data.info()
```

```python
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
data1 =data.fillna(method="bfill")
data1
```

```python
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
data1['POLIZA'] =data1['POLIZA'].fillna(method="ffill")
data1
```

```python
fig = plt.figure(figsize =(15, 8))
data1.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

##Obtenemos la desviación estándar para eliminar outliers en el df

```python
y=data1
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
data2= data1[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data2
```

```python
columnas_a_eliminar = ['FECHA', 'FOLIO', 'UUID', 'RFC', 'PROVEEDOR', 'TIPO GASTO', 'DESCRIPCION', 'TIPO','STATUS', 'POLIZA']
data2 = data1.drop(columnas_a_eliminar, axis=1)
data2
```

```python
data_clean=data2.copy()
data_clean=data_clean.fillna(round(data2.mean(),1))
data_clean
```

```python
valores_nulos = data_clean.isnull().sum()
valores_nulos
```

```python
fig2 = plt.figure(figsize =(15, 8))
data_clean.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe DE")
plt.show() #dibujamos el diagrama
```

```python
folio = data['FOLIO']
Datos_limpios = pd.concat([folio, data_clean], axis=1)
Datos_limpios
```

```python
Datos_limpios.to_csv("Datos_limpios.csv")
```
